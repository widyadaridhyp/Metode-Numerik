```python
import numpy as np #impor numpy yang nanti akan berguna untuk pengolahan matriks
from tabulate import tabulate #impor tabulate yang nanti akan berguna pada pembuatan tabel

while True: #Untuk melakukan pengulangan agar progam dapat diulang-ulang tanpa mengcopy
    opsi = int(input('Ingin melanjutkan program? (Ya = 1, Tidak = 2): '))
    if opsi == 1:
        aug_matrix = np.array(eval(input('Masukkan augmented matrixnya: '))) #input matriksnya
        initial_solution = np.array(eval(input('Masukkan tebakan awal nilainya: ')), dtype = float) #input nilai x0 atau nilai x untuk teban awal
        tolerance = eval(input('Masukkan toleransi: ')) #input toleransinya 
        
        table = [] #inisialkan matriks kosong yang nanti akan ditambahkan elemennya dari hasil iterasi gauss seidel
        
        opsi_method = int(input('Mau pake metode apa? 1 = Gauss Seidel, 2 = Jacobi, 3 = SOR: '))
        
        if opsi_method == 1:
            def GaussSeidel(matrix, initial, tol):
                n = np.shape(matrix)[0] #ukuran matriks/banyak variabel
                x0 = initial #definisikan x0 sebagai tebakan awal
                x = x0.copy()  #definisikan x = x0 (gunakan .copy() agar jika x0 berubah, x tidak berubah)
                a = 0
                for i in range (n):
                    b = matrix[i, n] #konstanta pada SPL
                    for j in range(n): #metode Gauss-Seidel
                        if j != i:
                            if j < i:
                                b -= matrix[i,j]*x[j]
                            else:
                                b -= matrix[i,j]*x0[j]
                    x[i] = b / matrix[i,i] #hasil aproksimasi Gauss-Seidel
        
                while np.linalg.norm(x-x0,np.inf) >= tol:  #pengecekan kriteria penyetop menggunakan error mutlak
                    x0 = x #definisikan x0 sebagai x(k-1)
                    x = x0.copy() #definisikan x sebagai x(k), namun awalnya samakan dengan x0
                    for i in range (n): #metode Gauss-Seidel
                        b = matrix[i, n]
                        for j in range(n):
                            if j != i:
                                if j < i:
                                    b -= matrix[i,j]*x[j]
                                else:
                                    b -= matrix[i,j]*x0[j]
                        x[i] = b / matrix[i,i]
                        entri = x.tolist()
                    a = a+1
                    entri.insert(0, a) #untuk menambahkan sebanyak a
                    table.append(entri) #untuk menambahkan himpunan kosong atau table tadi
                print(tabulate(table, headers = ["Iterasi", "x1", "x2", "x3", "x4"], tablefmt = "fancy_grid")) #Buat header tabel
                return x #nilai yang direturn fungsi ini adalah x(k)

            print ('Aproksimasi Solusi Gauss Seidel: ') 
            print(GaussSeidel(aug_matrix, initial_solution, tolerance)) #cetak nilai aproksimasinya
        
        if opsi_method == 2:
            def Jacobi(matrix,initial,tol):                
                n = np.shape(matrix)[0]                    #ukuran matriks/banyak variabel
                x0 = initial                               #definisikan x0 sebagai tebakan awal
                x = x0.copy()                              #definisikan x = x0 (gunakan .copy() agar 
                z = 0
                                                            #jika x0 berubah, x tiidak berubah)
                for i in range(n):
                    b = matrix[i,n]                        #konstanta pada SPL
                    for j in range(n):                     #metode Jacobi
                        if j!=i:
                            b-=matrix[i,j]*x0[j]           
                    x[i]=b/matrix[i,i]                     #hasil aproksimasi Jacobi
                    entri = x.tolist()
                z = z+1
                entri.insert(0, z)
                table.append(entri)
    
                while np.linalg.norm(x-x0,np.inf)>=tol:    #pengecekan kriteria penyetop menggunakan error mutlak
                    x0 = x                                 #definisikan x0 sebagai x(k-1)
                    x = x0.copy()                          #definisikan x sebagai x(k), namun awalnya samakan dengan x0
                    for i in range(n):                     #metode Jacobi
                        b = matrix[i,n]
                        for j in range(n):
                            if j!=i:
                                b-=matrix[i,j]*x0[j]
                        x[i]=b/matrix[i,i]
                        entri = x.tolist()
                    z = z+1
                    entri.insert(0, z)
                    table.append(entri)
                print(tabulate(table, headers = ["Iterasi", "x1", "x2", "x3", "x4"], tablefmt = "fancy_grid"))
                return x                                   #nilai yang direturn fungsi ini adalah x(k)

            print('Aproksimasi solusi SPL tersebut dengan metode Jacobi adalah :')
            print(Jacobi(aug_matrix,initial_solution,tolerance))
            
        if opsi_method == 3:
            w = eval(input('Masukkan faktor relaksasi : '))
            def SOR(matrix,initial,w,tol):                
                n = np.shape(matrix)[0]                    #ukuran matriks/banyak variabel
                x0 = initial                               #definisikan x0 sebagai tebakan awal
                x = x0.copy()
                z= 0                                       #definisikan x = x0 (gunakan .copy() agar 
                                                           #jika x0 berubah, x tidak berubah)
                for i in range(n):
                    b = matrix[i,n]                        #konstanta pada SPL
                    for j in range(n):                     #metode Gauss-Seidel
                        if j!=i:
                            if j<i:
                                b-=matrix[i,j]*x[j]
                            else:
                                b-=matrix[i,j]*x0[j]
                    x[i]=(1-w)*x0[i]+(b*w/matrix[i,i])     #hasil aproksimasi SOR
                    entri = x.tolist()
                z = z+1
                entri.insert(0, z)
                table.append(entri)
                
                while np.linalg.norm(x-x0,np.inf)>=tol:    #pengecekan kriteria penyetop menggunakan error mutlak
                    x0 = x                                 #definisikan x0 sebagai x(k-1)
                    x = x0.copy()                          #definisikan x sebagai x(k), namun awalnya samakan dengan x0
                    for i in range(n):                     #metode SOR
                        b = matrix[i,n]
                        for j in range(n):
                            if j!=i:
                                if j<i:
                                    b-=matrix[i,j]*x[j]
                                else:
                                    b-=matrix[i,j]*x0[j]
                        x[i]=(1-w)*x0[i]+(b*w/matrix[i,i])
                        entri = x.tolist()
                    z = z+1
                    entri.insert(0, z)
                    table.append(entri)
                print(tabulate(table, headers = ["Iterasi", "x1", "x2", "x3", "x4"], tablefmt = "fancy_grid"))
                return x                                   #nilai yang direturn fungsi ini adalah x(k)

            print('Aproksimasi solusi SPL tersebut dengan metode SOR adalah :')
            print(SOR(aug_matrix,initial_solution,w,tolerance))
    
    if opsi == 2:
        break
```

Program ini dibuat untuk membantu user mengerjakan permasalahan dengan bantuan Gauss Seidel dengan cara menginput matriks, tebakan awal, dan tolerance. Berikut algoritmanya.

3. Input matriks, tebakan awal, dan tolerance
4. Inisialkan table sebagai matriks kosong yang nantinya akan ditambahkan elemennya dengan cara mendefinisikan gauss seidel
5. Definisikan gauss seidel
6. Inisialkan banyak variabel, tebakan awal, dan x0.
7. Masuk ke dalam iterasi gauss seidel
8. Melakukan pengecekan kriteria penyetop menggunakan error mutlak
9. Penambahan elemen ke dalam matriks kosong dan pembuatan tabel
10. Cetak
